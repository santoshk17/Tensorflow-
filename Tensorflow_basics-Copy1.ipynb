```python
import tensorflow as tf
```

 
 let us take a look at a Python program which adds two numbers 

```python
def addition(x, y):
    z=x+y
    return z
```

now let's add two numbers a and b

```python
a=5
b=2
addition(a,b)
```

The python function is straightaway executed.

Unlike this, to execute the same in tensorflow, a graph has to be made. This graph can only be invoked through a tensorflow session. A tensorflow session allocates all the resources demanded by graph on cpu.  

```python
a = tf.placeholder(tf.float32, name="a")
b = tf.placeholder(tf.float32, name="b")
```

```python
c = tf.add(a, b, name="addition")
```

```python
from IPython.display import Image
Image("tfsess.JPG")
```

```python
with tf.Session() as session:

    result = session.run(addition, feed_dict={a: 2, b:4})

    print(result)
```
